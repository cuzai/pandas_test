```python
import matplotlib.pyplot as plt
```

```python
# 그려 낸 그래프들이 실시간으로 바로 화면에서 보일 수 있게 해 주는 것
%matplotlib inline
```

```python
import numpy as np
# 0부터 5 사이에 등간격으로 11개짜리 리스트를 생성해라
x = np.linspace(0, 5, 11)
x
```

```python
y = x **2
y
```

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
# 여러개 그래프를 생성
# plt.subplot(몇개 열, 몇개 행, 몇번째에 그래프를 노출시킬건지)

plt.subplot(1, 2, 1)
plt.plot(x,y)

plt.subplot(1, 2, 2)
plt.plot(y,x)
```

```python
# 비어있는 캔버스를 갖다놓고, 
fig = plt.figure()
```

```python
# 축을 정하고,
ax1 = fig.add_axes( [0.1, 0.1, 0.8, 0.8] )
ax2 = fig.add_axes( [0.2, 0.5, 0.4, 0.3] )
```

```python
# 그래프를 그린다
ax1.plot(x, y)
ax1.set_xlabel('X')
ax1.set_ylabel('y')
ax1.set_title('Title')

ax2.plot(y,x)
ax2.set_title('Smaller Plot')
fig
```

```python
fig2 = plt.figure(figsize = (8,4), dpi = 60)
ax = fig2.add_axes( [0,0,1,1] )
ax.plot(x,y)
ax.set_title('dd')
ax.set_xlabel('x')
ax.set_ylabel('y')
```

```python
fig.savefig('mypic.png', dpi = 200)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, x**2, label = 'X_squred')
ax.plot(x, x**3, label = 'X_cubed')
ax.legend()

ax.plot(x, x*2, color = 'black', linewidth = 5, alpha = 0.5, linestyle = '--', marker = 'o', markersize = 20, markerfacecolor = 'yellow', markeredgewidth = 3, markeredgecolor = 'green')
```

```python
# 그래프에서의 보여주고 싶은 범위를 지정. limit
ax.set_xlim([0,1])
ax.set_ylim([0,50])

fig
```
